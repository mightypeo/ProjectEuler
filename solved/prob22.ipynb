Using **0022_names.txt**a 46K text file containing over five-thousand first names, begin by sorting it into alphabetical order. Then working out the alphabetical value for each name, multiply this value by its alphabetical position in the list to obtain a name score.

For example, when the list is sorted into alphabetical order, COLIN, which is worth $3 + 15 + 12 + 9 + 14 = 53$, is the $938$th name in the list. So, COLIN would obtain a score of $938 \times 53 = 49714$.

What is the total of all the name scores in the file?


```python
ANSWER = 871198282
FILENAME = '0022_names.txt'
```

```python
%%time
import os
with open(os.path.join("..", FILENAME)) as f:
    content = f.read()
names = sorted(list(map( lambda x: x.strip('"'), content.split(','))))
```

```python
def name_score(position:int, name: str):
    return position *sum( ord(c) - ord('A') + 1 for c in name)
```

```python
%%time
name_scores = {}
for i, name in enumerate(names):
    name_scores[name] = name_score(i+1, name)
```

```python
%%time
sum(name_scores.values())
```

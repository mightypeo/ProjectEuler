<p>How many such routes are there through a $20 \times 20$ grid?</p>


```python
ANSWER = 137846528820
```

```python
def binomial(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
```

```python
ANSWER - binomial(40, 20)
```

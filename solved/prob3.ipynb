```python
factors = []
for i in range(2, math.isqrt(num)):
    if num % i == 0:
        print( f"{i}")
        factors.append(i)
```

```python
first_primes = [ 2, 3, 5, 7, 11, 13, 17, 19, 23]
def is_prime(n) -> bool:
    """ a prime number is only divisible by itself and 1"""
    for p in first_primes:
        if n % p == 0:
            return False
    for f in range(2,math.isqrt(n)):
        if n % f == 0:
            return False
    return True
```

```python
[ (n, is_prime(n)) for n in factors]
```

```python
prime_factors = [ n for n in factors if is_prime(n)]
```

```python
sum(prime_factors)
```

```python
max(prime_factors)
```

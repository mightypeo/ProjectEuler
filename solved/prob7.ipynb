What is the $10\,001$st prime number?

```python
def is_prime(n):
    # Prime numbers must be greater than 1
    if n <= 1:
        return False
    # 2 and 3 are prime
    if n <= 3:
        return True
    # Eliminate even numbers and multiples of 3
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    # Check factors from 5 up to the square root
    # Primes > 3 follow the pattern 6k ± 1
    for i in range(5, int(math.sqrt(n)) + 1, 6):
        if n % i == 0 or n % (i + 2) == 0:
            return False
            
    return True
```

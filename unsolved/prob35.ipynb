# Circular Primes

_Problem 35_

The number, $197$, is called a circular prime because all rotations of the digits: $197$, $971$, and $719$, are themselves prime.

There are thirteen such primes below $100$: $2, 3, 5, 7, 11, 13, 17, 31, 37, 71, 73, 79$, and $97$.

How many circular primes are there below one million?


```python
import sys
sys.path.append('..')
from common.common import first_primes, is_prime
```

```python
def is_circular_prime(n):
    s = str(n)
    for i in range(len(s)):
        if not is_prime(int(s[i:] + s[:i])):
            return False
    return True
```

```python
%%time
primes_by_rotation = {}
result = []
for i in range(100, 1_000_000):
    if is_prime(i):
        key = "".join(sorted(list(str(i))))
        if key in primes_by_rotation:
            primes_by_rotation[key].append(i)
            # result.append(i)
            continue
        if is_circular_prime(i):
            primes_by_rotation[key] = [i]
            result.append(i)

len(primes_by_rotation), len(result)
```
